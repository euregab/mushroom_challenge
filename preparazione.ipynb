# Notebook di preparazione del dataset

```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
path = './mushrooms.partial.csv'
data = pd.read_csv(path)
```

## Vediamo i valori delle colonne

```python
for col in data.columns:
    print(col)
    print(data[col].value_counts(), "\n")
```

I valori delle features corrispondono a quanto indicato sul moodle. Considero il valore '?' dello *stalk-root* come un valore reale (anche perché sono ben 2007 elementi ad avere '?' come stalk-root)

## Vediamo se ci sono valori nulli o mancanti

```python
for col in data.columns:
    print(col)
    print(data[col].isnull().sum(), "\n")
```

Non ci sono valori nulli o mancanti in nessuna colonna, pertanto in nessuna riga

## Split del dataset in train e test

La ragione scelta per lo split è 70-15-15.
Il 15% dei dati andrà nel file di test.
L'85% dei dati andrà nel file di training.
Pertanto nel file di training, la porzione di training sarà circa dell'82% mentre la porzione per il validation sarà del 18%

```python
train_data, test_data = train_test_split(data, random_state=10, test_size=0.15)
```

```python
path_train = './train.csv'
path_test = './test.csv'

train_data.to_csv(path_train, index=False)
test_data.to_csv(path_test, index=False)
```
